```python
import pandas as pd


data = pd.read_csv("C:/Users/gowth/Downloads/autos.csv", encoding='ISO-8859-1')
```

```python
# Check for missing values
print(data.isnull().sum())

# Display data types of each column
print(data.dtypes)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
data = data.dropna(subset=['vehicleType', 'gearbox', 'model', 'fuelType', 'notRepairedDamage'])
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Convert InvoiceDate to datetime
data['dateCrawled'] = pd.to_datetime(data['dateCrawled'])
```

```python
print(data.describe())
```

```python
print(data.head())

# Check for missing values
print(data.isnull().sum())

# Convert InvoiceDate to datetime, ensuring the conversion is successful
data['dateCrawled'] = pd.to_datetime(data['dateCrawled'], errors='coerce')

# Verify the conversion
print(data['dateCrawled'].dtypes)

# Check for duplicate rows
duplicate_rows = data[data.duplicated()]
print("Number of duplicate rows:", duplicate_rows.shape[0])

```

```python
data['price'] = data['price'].str.replace('[\$,]', '').astype(float)
```

```python
data['odometer'] = data['odometer'].str.replace('km', '').str.replace(',', '').astype(float)
```

```python
data['dateCrawled'] = pd.to_datetime(data['dateCrawled'])
data['dateCreated'] = pd.to_datetime(data['dateCreated'])
data['lastSeen'] = pd.to_datetime(data['lastSeen'])
```

```python
categorical_columns = ['name', 'seller', 'offerType', 'abtest', 'vehicleType', 'gearbox', 
                       'model', 'fuelType', 'brand', 'notRepairedDamage']
for col in categorical_columns:
    data[col] = data[col].astype('category')
```

```python
data['vehicleType'] = data['vehicleType'].str.lower().str.strip()
data['gearbox'] = data['gearbox'].str.lower().str.strip()
```

```python
data = data[(data['price'] > 0) & (data['price'] < data['price'].quantile(0.99))]
data = data[(data['odometer'] > 0) & (data['odometer'] < data['odometer'].quantile(0.99))]
```

```python
print(data.dtypes)
print(data.isnull().sum())
```

```python
print(data.head())
```

```python
# Get an overview of the dataset
print("Shape of the dataset:", data.shape)
print("Data types and missing values:\n", data.info())
print("Summary of missing values:\n", data.isnull().sum())
```

```python
# Descriptive statistics for numerical columns
print("Descriptive statistics for numerical columns:\n", data.describe())

# Descriptive statistics for categorical columns
print("Descriptive statistics for categorical columns:\n", data.describe(include=['category']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(y='vehicleType', data=data, order=data['vehicleType'].value_counts().index)
plt.title('Count of Vehicle Types')
plt.xlabel('Count')
plt.ylabel('Vehicle Type')
plt.show()
```

```python
# Price vs. Year of Registration
plt.figure(figsize=(10, 6))
sns.scatterplot(x='yearOfRegistration', y='price', data=data)
plt.title('Price vs. Year of Registration')
plt.xlabel('Year of Registration')
plt.ylabel('Price')
plt.show()

# Boxplot of Price by Vehicle Type
plt.figure(figsize=(12, 8))
sns.boxplot(x='vehicleType', y='price', data=data)
plt.title('Boxplot of Price by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
